```python
class Product():     
    def __init__(s,name,name1,name2):
        s.name=name
        s.name1=name1
        s.name2=name2
        s.__tquantity=0
        s.__squantity=0
        s.__jquantity=0
        s.__total=0
        s.__tshirtcost=0
        s.__shirtcost=0
        s.__jeanscost=0
        
    def tshirt(s):
        s.__tquantity=int(input("How much tshirt you want to buy ? If not want to buy Enter 0"))
        s.__total=s.__total+(s.__tquantity*250)
        s.__tshirtcost=s.__tshirtcost+(s.__tquantity*250)
        print(f'Total cost of {s.__tquantity} {s.name} is {s.__tshirtcost}\n')
        
    def shirt(s):
        s.__squantity=int(input("How much shirt you want to buy ? If not want to buy Enter 0"))
        s.__total=s.__total+(s.__squantity*500)
        s.__shirtcost=s.__shirtcost+(s.__squantity*500)
        print(f'Total cost of {s.__squantity} {s.name1} is {s.__shirtcost}\n')
        
    def jeans(s):
        s.__jquantity=int(input("How much jeans you want to buy ?If not want to buy Enter 0"))
        s.__total=s.__total+(s.__jquantity*800)
        s.__jeanscost=s.__jeanscost+(s.__jquantity*800)
        print(f'Total cost of {s.__jquantity} {s.name2} is {s.__jeanscost}\n')
      
    def bill(s):
        print(f'Your Final bill is {s.__total}\n')
        
    def display(s):
        print("Tshirt = 250 \nShirt = 500 \nJeans = 800\n")
        
    def overview(s):
        print("You have purchased")
        print(f'{s.__tquantity} {s.name} for {s.__tshirtcost}/-')
        print(f'{s.__squantity} {s.name1} for {s.__shirtcost}/-')
        print(f'{s.__jquantity} {s.name2} for {s.__jeanscost}/-')
        print("\n")
        
p=Product("Tshirt","Shirt","Jeans")
print("welcome to shopping Portal")
p.display()
p.overview()
while True:
    
    select=int(input("Press 1 to buy \nPress 2 to bill \nPress 3 to exit\n"))
    if select==1:
        choice=int(input("Press 1 for tshirt \nPress 2 for Shirt \nPress 3 for Jeans\n\n"))
        if choice==1:
            p.tshirt()
            p.overview()
        
        elif choice==2:
            p.shirt()
            p.overview()
        
        elif choice==3:
            p.jeans()
            p.overview()
        
    elif select==2:
        p.overview()
        p.bill()
        break
        
    elif select==3:
        break
```
